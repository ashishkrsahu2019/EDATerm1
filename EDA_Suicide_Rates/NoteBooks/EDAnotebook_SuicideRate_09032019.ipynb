### 3.3 Preprocessing

- Dealing with missing values<br/>
    - Replacing missing values of __HDI for year__ with median values.Assuming that HDi for different countries will vary, so we will replace any null value in this column with median calculated for that country.
    - 

```python
## Fix incorrect country name
data_df = data_df.replace('Republic of Korea', 'South Korea')
data_df = data_df.replace('Russian Federation', 'Russia')
```

```python
## Unique list of cars
myList= list(data_df['country'].unique())
```

```python
'''
This method is written to pre-process, 
replace NaN values in a data frame having integer data type with median, groupby country.
'''
def preProcess(df,lst,colName):
    for x in lst:
        my_med=df[df.country==x][colName].median()
        df[colName].fillna(my_med, inplace = True) ### replace colName by median value
```

```python
'''
This method is written to pre-process, replace NaN values in a data frame having integer data type with median.
Based on the car type.
'''
def preProcessReplaceZero(df,lst,colName):
    for x in lst:
        my_med=df[df.car==x][colName].median()
        df[colName].replace(0, my_med, inplace = True) 
```

 - Replacing missing values of __HDI for year__ with median values.Assuming that HDI for different countries will vary, so we will replace any null value in this column with median calculated for that country.

```python
## Replacing missing values of HDI, country wise.
preProcess(data_df,myList,'HDI for year')
```

```python
## Validate if all the Nan values have been addressed.
data_df.isnull().sum()
```

```python
## Save data frame for future use
data_df.to_csv('SucideRate_Post_Profile.csv')
```

### 4.13 Establish coorelation between all the features using heatmap.

```python
corr = data_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,vmax=.8,linewidth=.01, square = True, annot = True,cmap='YlGnBu',linecolor ='black')
plt.title('Correlation between all the parameters')
```

#### Observation

- __gdp_per_capita ($)__ and __HDI for year__ have good positive co-relation.
- __suicide_no__ and __population__ have good positive co-relation.
- __year__ and __suicide_no__ have negative co-relation.
- __year__ and __suicide/100k pop__ have negative co-relation.
- This shows that GDP per capita is directly effected by increase in HDI, and number of suicide is directly proportional to population.
